```python
import pandas as pd

def extract_first_two_words(text):
    words = text.split()
    return ' '.join(words[:1]) if len(words) >= 1 else text

def filter_and_print_unique_reviews(df_pmd):
    # Extract the first two words from each review
    df_pmd['first_two_words'] = df_pmd['review'].apply(extract_first_two_words)
    # Count the occurrences of each unique statement based on the first two words
    statement_counts = df_pmd['first_two_words'].value_counts()
    # Print the first unique statement and its count
    first_statement = statement_counts.index[31]
    first_statement_count = statement_counts.iloc[31]
    print(f"First unique statement: {first_statement} (Count: {first_statement_count})")

    return statement_counts

def main():


    # Get the count of unique reviews
    statement_counts = filter_and_print_unique_reviews(df_pmd)

    # Print the total number of unique reviews
    print(f"Total number of unique reviews: {len(statement_counts)}")

if __name__ == "__main__":
    main()
```

```python

df = pd.read_csv('filtered_data_token_length.csv')
# Step 1: Filter rows matching the condition 
filtered_df = df[(df['tool'] == 'PMD') & df['review'].str.startswith('Avoid excessively')]
print(f'Number of filtered rows: {len(filtered_df)}')

# Step 2: Remove duplicates based on 'Index' in the filtered DataFrame
filtered_df_no_duplicates = filtered_df.drop_duplicates(subset='Index')

# Step 3: Sample 20% of these filtered rows
sampled_df = filtered_df_no_duplicates.sample(frac=0.3, random_state=1)
print(f'Number of sampled rows: {len(sampled_df)}')

# Step 4: Filter rows from df that do not match the condition and have 'tool' as 'PMD'
non_matching_df = df[(~df['review'].str.match(r'^Avoid excessively\b', case=False, na=False))]

# Step 5: Combine sampled_df with non_matching_df
df = pd.concat([sampled_df, non_matching_df])

# Optional: Reset index if needed
df = df.reset_index(drop=True)

# Print the final DataFrame and its length
print(f'Number of rows in result_df: {len(df)}')
df.to_csv('filtered_data_token_length.csv', index=False)
```
